# Central Limit Theorem (CLT)

The Central Limit Theorem is one of the most important concepts in statistics.

It states that when we take many samples from a population and compute their means,
the distribution of those sample means will approach a **normal distribution**
as the sample size increases — even if the original population is not normally distributed.

In this notebook we will:

1. Create a non-normal population distribution
2. Draw repeated samples from that population
3. Compute the mean of each sample
4. Visualize how the distribution of sample means becomes approximately normal

This concept is the foundation for many statistical methods such as
confidence intervals, hypothesis testing, and A/B testing.

```python
import sys
import os
sys.path.append(os.path.abspath(".."))

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.utils import sample_means, plot_histogram, set_seed

set_seed(42)
```

## Generate a Skewed Population

To demonstrate the Central Limit Theorem, we start with a population
that is **not normally distributed**.

We use an **exponential distribution**, which is strongly skewed.

```python
population = np.random.exponential(scale=2,size = 100000)
```

### Population Distribution

```python
plt.figure(figsize=(5,4))
plot_histogram(data = population,title="Population Distribution (Skewed)",xlabel="Value")
plt.show()   
```

## Sampling from the Population

We repeatedly draw samples from the population and compute
the mean of each sample.

We will compare different sample sizes to see how the distribution
of sample means changes.

```python
sample_sizes = [5, 30, 100]
plt.figure(figsize=(18,4))
#first subplot : raw population
plt.subplot(1,len(sample_sizes)+1,1)
plot_histogram(population,xlabel="Value",title="Raw Population (Skewed)")
#second the sample means plots 
for i , size in enumerate (sample_sizes,2):
    means = sample_means(population,size,n_samples=1000)
    plt.subplot(1,len(sample_sizes)+1,i)
    plot_histogram(means,xlabel="Mean Value",title = f"Sample Size = {size}")
    
    
plt.tight_layout()
plt.show()    
```

## Interpretation

The population distribution is highly skewed.

However, as we increase the sample size, the distribution
of the sample means becomes more symmetric and begins
to resemble a normal distribution.

This illustrates the **Central Limit Theorem**:

Even when the population distribution is not normal,
the distribution of sample means approaches a normal
distribution as the sample size increases.

## Key Takeaways

- The Central Limit Theorem explains why normal distributions
  appear frequently in statistics.
  
- The sampling distribution of the mean becomes approximately
  normal as the sample size increases.

- This property allows statisticians and data scientists to use
  normal-based inference methods such as:
  
  - Confidence intervals
  - Hypothesis testing
  - A/B testing
